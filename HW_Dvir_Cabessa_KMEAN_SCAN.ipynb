```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans, DBSCAN
```

```python
X1, _ = make_moons(n_samples=300, noise=0.05, random_state=0)
X2, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

```python
def plot_clusters(X, labels, title, ax):
    ax.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=10)
    ax.set_title(title)
```

```python
kmeans_moons = KMeans(n_clusters=2, random_state=0).fit(X1)
kmeans_blobs = KMeans(n_clusters=3, random_state=0).fit(X2)
kmeans_blobs
```

```python
dbscan_moons = DBSCAN(eps=0.2, min_samples=5).fit(X1)
dbscan_blobs = DBSCAN(eps=0.3, min_samples=5).fit(X2)
dbscan_blobs
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

plot_clusters(X1, dbscan_moons.labels_, "DBSCAN - Moons", axes[0, 0])
plot_clusters(X1, kmeans_moons.labels_, "KMeans - Moons", axes[0, 1])

plot_clusters(X2, dbscan_blobs.labels_, "DBSCAN - Blobs", axes[1, 0])
plot_clusters(X2, kmeans_blobs.labels_, "KMeans - Blobs", axes[1, 1])

plt.tight_layout()
plt.show()
```
