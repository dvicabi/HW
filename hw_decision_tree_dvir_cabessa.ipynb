```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    'Discount': ['Yes', 'Yes', 'No', 'No', 'Yes', 'No'],
    'Purchase': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No'],
    'Returned': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No']
})
```

```python
X = pd.get_dummies(data[['Discount', 'Purchase']], drop_first=True)
y = data['Returned'].map({'Yes': 1, 'No': 0})
y
```

```python
clf_gini = DecisionTreeClassifier(criterion='gini', random_state=0)
clf_gini.fit(X, y)
```

```python
new_customer = pd.DataFrame(columns=X.columns)
new_customer.loc[0] = [1, 0]
prediction = clf_gini.predict(new_customer)
prediction
```

```python
plt.figure(figsize=(10, 6))
plot_tree(clf_gini, feature_names=X.columns, class_names=["No", "Yes"], filled=True)
plt.tight_layout()
plt.show()
```

```python
clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state=0)
clf_entropy.fit(X, y)
```

```python
plt.figure(figsize=(10, 6))
plot_tree(clf_entropy, feature_names=X.columns, class_names=["No", "Yes"], filled=True)
plt.tight_layout()
plt.show()
```
